```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import fsolve

#1.2
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(131, projection = '3d')
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = -X**2 + X*Y + Y**2
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

#1.3
def equations(X):
    return X**2 + X - 1 + 3*X + 1
x_intersect = fsolve(equations, [0, -1])
y_intersect = x_intersect**2 + x_intersect - 1
ax2 = fig.add_subplot(132)
x = np.linspace(-3, 2, 400)
f_x = x**2 + x -1
g_x = -3*x - 1
ax2.plot(x, f_x, label='f(x) = x^2 + x - 1', color='blue')
ax2.plot(x, g_x, label='g(x) = -3x - 1', color='red')
ax2.scatter(x_intersect, y_intersect, color='black', zorder=3)
for xi, yi in zip(x_intersect, y_intersect):
    ax2.text(xi, yi, f'({xi:.2f}, {yi:.2f})', fontsize=10, verticalalignment='bottom')
ax2.legend()
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Function Intersection')
ax2.grid()

#1.4
ax3 = fig.add_subplot(133)
scores = [76, 10, 21, 61, 40, 65, 40, 95, 77, 92, 59, 81, 76, 98, 57, 51, 92, 35, 17, 61, 35]
ax3.hist(scores, bins=np.linspace(0, 100, 6), edgecolor='black', alpha=0.75)
ax3.set_xlabel('Score Ranges')
ax3.set_ylabel('Frequency')
ax3.set_title('Histogram of Student Scores')
ax3.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```
